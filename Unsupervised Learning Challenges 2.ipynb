Let's name the clusters 2:

Calculate the tf-idf of each word in each cluster (think of all sentences of a cluster together as a document). Represent each cluster with the top 1, or top 2 or... to 5 tf-idf words. For each cluster, print the name (keywords) of the cluster, and "N statements" in the cluster (N is the size of the cluster)

```python
##Function to combine information into clustered data
def clustered_combined_func(estimator, n_clusters, n_init, vectors, plainText, n):

    classifier = estimator(n_clusters=n_clusters, n_init=n_init)
    model_current = classifier.fit(vectors)
    
    labels = model_current.labels_
    cluster_list = model_current.cluster_centers_
    cluster_lengths = Counter(model_current.labels_) 
    
    clustered_text=[x for (y,x) in sorted(zip(labels, plainText))]
    
    top_clusters = sorted(zip(cluster_lengths.values(), cluster_lengths.keys()), reverse = True)
    top_cluster_list = []
    sorted_labels = sorted(labels)

    N = n
    clustered_combined_data= {}

    for i in range(N):
        clustered_combined_data[i]=''
    for i in range(len(labels)):
        clustered_combined_data[sorted_labels[i]] += clustered_text[i]
    
    listname = clustered_combined_data.values()
    
    return listname, cluster_lengths
```

###Challenge 6

Cluster the same data with MiniBatchKMeans. MiniBatchKMeans is a fast way to apply K-means to large data without much loss -- The results are very similar. Instead of using EVERY single point to find the new place of the centroid, MiniBatch just randomly samples a small number (like 100) in the cluster to calculate the new center. Since this is usually very close to the actual center, the algorithm gets there much faster. Try it and compare the results.



```python
from sklearn.cluster import MiniBatchKMeans

##Using 10 clusters
kminimeans_classifier = MiniBatchKMeans(n_init=7, n_clusters=10)

kmini_model = kminimeans_classifier.fit(tweet_doc_vectors)
label = kmini_model.labels_

clustered_tweets=[x for (y,x) in sorted(zip(label, tweet_text))]
labels = sorted(label)

print "INERTIA:", kmini_model.inertia_

for i in range(len(labels)):
    print labels[i], clustered_tweets[i]
```
